# 4.3.1 例題: 集約されたデータの扱い

```julia
using Distributions
using Plots
```

```julia
std_height = 15
mean_height = 150

n_counties = 5000
population = rand.(DiscreteUniform(100, 1500), n_counties)
```

```julia
average_across_county = mean.(rand.(Normal(mean_height, std_height), population))
i_min = argmin(average_across_county)
i_max = argmax(average_across_county)
```

```julia
scatter(population[[i_min, i_max]], average_across_county[[i_min, i_max]],
    label = "Extreme heights",
    markersize = 10, markershape = :diamond)
scatter!(population, average_across_county,
    label = "",
    markersize = 2)
hline!([mean_height],
    linestyle = :dash, linewidth = 3, linecolor = :lime,
    label = "True expected height",
    xlabel = "County population", ylabel = "Average height in county")
xlims!(minimum(population), maximum(population))
```

Population sizes of 10 shortest countries

```julia
population[sortperm(average_across_county)[1:10]]
```

Population sizes of 10 tallest countries

```julia
population[sortperm(average_across_county, rev = true)[1:10]]
```
