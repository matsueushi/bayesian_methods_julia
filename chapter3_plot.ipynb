# 3.1 山あり谷あり、分布の地形

```julia
using Distributions
using Plots
```

```julia
x = y = 0:0.5:5
```

```julia
uniform = Uniform(0, 5)
z = Surface((s, t) -> pdf(uniform, s) * pdf(uniform, t), x, y)
surface(x, y, z, camera=(80, 30), clim = (0, 0.05))
```

```julia
heatmap(x, y, z, aspect_ratio = 1, clim = (0, 0.05))
```

```julia
exp_x = Exponential(3)
exp_y = Exponential(10)
z = Surface((s, t) -> pdf(exp_x, s) * pdf(exp_y, t), x, y)
surface(x, y, z, linealpha = 0.1, camera=(80, 30), clim = (0, 0.05))
```

```julia
heatmap(x, y, z, aspect_ratio = 1, clim = (0, 0.05))
```
